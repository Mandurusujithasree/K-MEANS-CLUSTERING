# Importing Libraries

```python
from sklearn .datasets import make_blobs #make_blobs is used for artifical data.
```

```python
raw_data = make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std = 1.8) 
```

```python
import pandas as pd #used for data analysis.

import numpy as np #mathematical operatorions on multi-dimensional arrays.

import seaborn #statistical graphics

import matplotlib.pyplot as plt #used for data visualization

%matplotlib inline
```

# Explanatory Data Analysis

```python
plt.scatter(raw_data[0][:,0], raw_data[0][:,1], c=raw_data[1]) #plot all of the rows in the first column of our data set against all of the rows in the second column of our data set
```

# K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=4) #declares the clusters count  
```

```python
model.fit(raw_data[0])
```

```python
model.labels_ #gives that datapoint belongs to one particular cluster
```

```python
model.cluster_centers_ #gives the cluster centre
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))

ax1.set_title('Our Model') #setting alarm for the first axis

ax1.scatter(raw_data[0][:,0], raw_data[0][:,1],c=model.labels_) #plots the graph for the model lables_

ax2.set_title('Original Data')

ax2.scatter(raw_data[0][:,0], raw_data[0][:,1],c=raw_data[1]) #plots the original data
```
